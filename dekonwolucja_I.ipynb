```python
from deconvolution import Deconvolution
#import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open('biop.tif')  
```

```python
decimg = Deconvolution(image=img, basis=[[0.65, 0.70, 0.29], [0.07, 0.99, 0.11]])
layer1, layer2 = decimg.out_images(mode=[1, 2])
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(25, 25))

ax = axes.ravel()
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3)

ax[0].imshow(img)
ax[0].set_title('Obraz oryginalny')
ax[0].axis('off')

ax[1].imshow(layer1)
ax[1].set_title('Pierwsze barwienie')
ax[1].axis('off')

ax[2].imshow(layer2)
ax[2].set_title('Drugie barwienie')
ax[2].axis('off')
```
